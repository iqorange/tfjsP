```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
```

```python
x
```

```python
y = np.sin(x)
y
```

```python
plt.plot(x, y)
```

```python
plt.show()
```

```python
cosy = np.cos(x)
```

```python
cosy.shape
```

```python
siny = y.copy()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red")
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red", linestyle="--")
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red", linestyle="--")
plt.xlim(-5, 15)
plt.ylim(0, 1.5)
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red", linestyle="--")
plt.axis([-1, 11, -2, 2])
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red", linestyle="--")
plt.xlabel("x axis")
plt.ylabel("y value")
plt.show()
```

```python
plt.plot(x, siny, label="sin(x)")
plt.plot(x, cosy, color="red", linestyle="--", label="cos(x)")
plt.xlabel("x axis")
plt.ylabel("y value")
plt.legend()
plt.show()
```

```python
plt.plot(x, siny, label="sin(x)")
plt.plot(x, cosy, color="red", linestyle="--", label="cos(x)")
plt.xlabel("x axis")
plt.ylabel("y value")
plt.legend()
plt.title("Welcome to the ML World~")
plt.show()
```

```python
plt.scatter(x, siny)
plt.show()
```

```python
plt.scatter(x, siny)
plt.scatter(x, cosy, color="red")
plt.show()
```
